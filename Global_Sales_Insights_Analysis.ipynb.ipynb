```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df.groupby('COUNTRY')['SALES'].sum().sort_values().plot(kind='bar')

plt.title("Sales by Country")
plt.xlabel("SALES")
plt.ylabel("COUNTRY")
plt.show()
```

```python
df.groupby('YEAR_ID')['SALES'].sum().plot(kind='bar')

plt.title("Sales by Year")
plt.xlabel("YEAR")
plt.ylabel("SALES")

plt.show()
```

```python
df.groupby('PRODUCTLINE')['SALES'].sum().plot(kind='bar')

plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# Insights

#1. USA contributes the highest sales among all countries, making it the most important market.

#2. Sales show variation across years, indicating growth and fluctuations in business performance.

#3. Certain product lines generate significantly higher revenue compared to others.

#4. Countries like France, UK, and Germany also contribute strong sales after USA.

#5. Higher quantity ordered generally leads to higher sales, showing a positive relationship.

#6. The heatmap shows a strong correlation between SALES and QUANTITYORDERED.

#7. Some columns have weak or no correlation, meaning they do not strongly affect sales.

#8. Business performance depends on both product type and region.
```
